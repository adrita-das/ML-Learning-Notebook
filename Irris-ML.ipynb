23. Detect outliers in petal width using the interquartile range method and count how many are present.

```python
q1=df1['petal_width'].quantile(0.25)
q3=df1['petal_width'].quantile(0.75)
iqr=q3-q1
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
outliers=df1[(df1['petal_width']<lower_bound) | (df1['petal_width']>upper_bound)]
outliers
```

```python
outlier_count = len(outliers)
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {iqr}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"\nNumber of Outliers in petal_width: {outlier_count}")

if outlier_count > 0:
    print("\nOutlier data points:")
    print(outliers)
else:
    print("\nNo outliers detected in petal_width.")
```
